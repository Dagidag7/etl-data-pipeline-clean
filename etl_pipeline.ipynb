```python
import pandas as pd
```

```python
orders = pd.read_csv("orders.csv")
products = pd.read_csv("products.csv")
aisles = pd.read_csv("aisles.csv")
departments = pd.read_csv("departments.csv")
order_products__prior = pd.read_csv("order_products__prior.csv")
order_products__train = pd.read_csv("order_products__train.csv")

print("Datasets Loaded")
```

```python
orders.head()
```

```python
products_full = products.merge(aisles, on="aisle_id")
products_full = products_full.merge(departments, on="department_id")

products_full.head()
```

```python
order_products = pd.concat([order_products__prior, order_products__train])

order_products.shape
```

```python
order_details = order_products.merge(products_full, on="product_id")

final_data = order_details.merge(orders, on="order_id")

final_data.head()
```

```python
final_data["product_name"].value_counts().head(10)

final_data.groupby("department")["order_id"].count()

final_data["user_id"].value_counts().head(10)
```

```python
from sqlalchemy import create_engine

engine = create_engine(
"postgresql://postgres:password@localhost:5432/etl_project"
)

final_data.to_sql(
"instacart_orders",
engine,
if_exists="replace",
index=False
)

print("Data Loaded into PostgreSQL")
```
